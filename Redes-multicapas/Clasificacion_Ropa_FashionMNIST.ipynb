#  Clasificaci贸n de Ropa (Fashion-MNIST)
Este notebook entrena una CNN para reconocer diferentes tipos de prendas del dataset **Fashion-MNIST**.

```

import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# Cargamos los datos
(X_train, y_train), (X_test, y_test) = datasets.fashion_mnist.load_data()

# Normalizamos y damos forma
X_train, X_test = X_train / 255.0, X_test / 255.0
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

# Nombres de las clases
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

## З Construcci贸n del modelo CNN
Usamos m谩s filtros para mejorar el reconocimiento visual.

```

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.summary()
```

## М Entrenamiento del modelo

```

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=7, validation_data=(X_test, y_test))
```

##  Resultados y evaluaci贸n

```

plt.plot(history.history['accuracy'], label='Precisi贸n de entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisi贸n de validaci贸n')
plt.legend()
plt.show()

test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print('\nPrecisi贸n final en datos de prueba:', test_acc)
```
