# М Clasificaci贸n de D铆gitos (MNIST)
En este notebook entrenamos una red neuronal convolucional (CNN) para reconocer d铆gitos escritos a mano usando el conjunto de datos **MNIST**.

```

# Importamos las librer铆as necesarias
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# Cargamos el dataset MNIST
(X_train, y_train), (X_test, y_test) = datasets.mnist.load_data()

# Normalizamos los datos (de 0 a 1)
X_train, X_test = X_train / 255.0, X_test / 255.0

# Cambiamos la forma a (28,28,1) para usar con CNN
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

print('Forma de los datos de entrenamiento:', X_train.shape)
```

##  Creaci贸n del modelo CNN
La red tendr谩 dos capas convolucionales seguidas de pooling, luego se aplana y se conecta a capas densas.

```

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.summary()
```

## 锔 Compilaci贸n y entrenamiento del modelo

```

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

##  Evaluaci贸n del modelo

```

plt.plot(history.history['accuracy'], label='Precisi贸n de entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisi贸n de validaci贸n')
plt.legend()
plt.show()

test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print('\nPrecisi贸n final en datos de prueba:', test_acc)
```
